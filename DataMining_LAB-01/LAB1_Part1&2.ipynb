```python
# Function to detect and replace outliers using IQR method
def detect_and_replace_outliers_IQR(df2, feature):
    Q1 = df2[feature].quantile(0.25)
    Q3 = df2[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"\nFor {feature}:")
    print(f"Q1 (25th percentile): {Q1}")
    print(f"Q3 (75th percentile): {Q3}")
    print(f"IQR: {IQR}")
    print(f"Lower bound for outliers: {lower_bound}")
    print(f"Upper bound for outliers: {upper_bound}")

    # Replace outliers with the median value of the feature
    median_value = df2[feature].median()

    # Apply replacement for outliers
    df2[feature] = df2[feature].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)

    return df2

# List of numeric features to check for outliers
features_to_check = ['Age', 'YearlyIncome', 'TotalChildren', 'NumberChildrenAtHome', 'NumberCarsOwned']

# Apply the outlier detection and replacement for each feature
for feature in features_to_check:
    df2 = detect_and_replace_outliers_IQR(df2, feature)
```
